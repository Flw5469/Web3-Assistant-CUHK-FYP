```python
from GoogleNews import GoogleNews
content_urls = []
googlenews = GoogleNews()
googlenews.search('crypto')
for i in range(1,10):
  content_urls+=googlenews.page_at(i)

googlenews.search('Caradano news')
for i in range(1,10):
  content_urls+=googlenews.page_at(i)
```
