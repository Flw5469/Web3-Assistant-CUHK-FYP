```python
def extract_strings(input_string):
    # Find the index of the first parenthesis
    left_paren_index = input_string.find('(')
    
    # Find the index of the first hyphen
    hyphen_index = input_string.find(') -')
    
    # Extract the substring before the first parenthesis
    before_paren = input_string[:left_paren_index].strip() if left_paren_index != -1 else input_string.strip()
    
    # Extract the substring after the first hyphen
    after_hyphen = input_string[hyphen_index + 3:].strip() if hyphen_index != -1 else ""
    
    return before_paren, after_hyphen

lines = text.split('\n')
before_list = []
after_list = []
for line in lines:
  before, after = extract_strings(line)
  before_list.append(before)
  after_list.append(after)


print(before_list)
print(after_list)
```
